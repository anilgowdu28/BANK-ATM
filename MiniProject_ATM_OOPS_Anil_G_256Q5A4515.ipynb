MINI PROJECT – PYTHON (**OOPS**) ATM MANAGEMENT SYSTEM


```python
from abc import ABC, abstractmethod
# Abstract Class
class BankAccount(ABC):
    @abstractmethod
    def transaction(self):
        pass
# Base Class (Encapsulation)
class Account(BankAccount):
    def __init__(self, name, acc_no, balance):
        self.name = name
        self._acc_no = acc_no
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    def set_balance(self, amount):
        self.__balance = amount
# Child Class (Inheritance)
class ATM(Account):
    @classmethod
    def bank_name(cls):
        print("\nWELCOME TO KIET BANK ATM")
    @staticmethod
    def valid_amount(amount):
        return amount > 0
    # Method Overriding
    def transaction(self):
        print("ATM Transaction Started")
    # Instance Methods
    def check_balance(self):
        print("Balance: ₹", self.get_balance())
    def deposit(self, amount):
        if self.valid_amount(amount):
            self.set_balance(self.get_balance() + amount)
            print("Deposit Successful")
    def withdraw(self, amount):
        if amount <= self.get_balance() and self.valid_amount(amount):
            self.set_balance(self.get_balance() - amount)
            print("Withdrawal Successful")
        else:
            print("Invalid Amount")
    def account_details(self):
        print("Name:", self.name)
        print("Account No:", self._acc_no)
        print("Balance: ₹", self.get_balance())
# Main Program
atm = ATM("ANIL", 4515, 5000)
ATM.bank_name()
atm.transaction()
while True:
    print("\n1.Check Balance")
    print("2.Deposit")
    print("3.Withdraw")
    print("4.Account Details")
    print("5.Exit")
    ch = input("Enter choice: ")
    if ch == '1':
        atm.check_balance()
    elif ch == '2':
        amt = int(input("Enter amount: "))
        atm.deposit(amt)
    elif ch == '3':
        amt = int(input("Enter amount: "))
        atm.withdraw(amt)
    elif ch == '4':
        atm.account_details()
    elif ch == '5':
        print("Thank You!")
        break
    else:
        print("Invalid Choice")
```

1. The program uses **OOPS concepts** to create a simple ATM system in Python.
2. "BankAccount" is an **abstract class** that defines a "transaction()"   method.
3. "Account" class stores account details using **encapsulation** (public, protected, private variables).
4. The balance is kept **private** and accessed using getter and setter methods.
5. "ATM" class **inherits** from "Account" and provides ATM functions.
6. The "transaction()" method is **overridden** in the ATM class.
7. A class method displays the bank name for all users.
8. A static method checks whether the entered amount is valid.
9. Instance methods allow the user to (check balance, deposit, and withdraw money.)
10. A while loop menu runs continuously until the user chooses Exit.
